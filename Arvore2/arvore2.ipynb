```python
import pandas as pd
from tabulate import tabulate

df = pd.read_csv(caminho_do_csv)

def arvore_decisao():
    def filtrar_dataframe(filtro, coluna):
        return df[df[coluna].str.contains(filtro, case=False, na=False)]

    def aplicar_intervalos(df, coluna, valor):
        df_copy = df.copy()
        
        # Tentando converter a coluna para valores num√©ricos, e gerando erro 'coerce' em caso de falha.
        df_copy[coluna] = pd.to_numeric(df_copy[coluna], errors='coerce')
        
        
        if isinstance(valor, (int, float)):
            return df_copy[(df_copy[coluna] >= valor - 3) & (df_copy[coluna] <= valor + 3)]
        else:
            # Se o valor n√£o for num√©rico, apenas retornar o dataframe sem filtro
            print(f"Valor inv√°lido para a coluna {coluna}, apenas valores num√©ricos s√£o aceitos.")
            return df_copy

    def jogadores_proximos(df, criterios, posicao_filtro):
        df_posicao = df[df['positions'].str.contains(posicao_filtro, case=False, na=False)]
        df_posicao = df_posicao.copy()
        df_posicao['distancia'] = df_posicao.apply(
            lambda row: sum(abs(row[crit] - val) for crit, val in criterios.items() if pd.notnull(row[crit])),
            axis=1
        )
        return df_posicao.sort_values(by='distancia').head(10)

    posicao = input("Qual posi√ß√£o geral deseja (Defensor, Atacante, Meia)? ").strip().capitalize()
    posicao_filtro = ""

    if posicao == "Defensor":
        tipo_defensor = input("Qual tipo de defensor (GK, CB, RB, LB)? ").strip().upper()
        posicao_filtro = tipo_defensor
        if tipo_defensor == "CB":
            altura = input("Qual √© a altura m√≠nima em cm? ").strip()
            heading_accuracy = input("Qual √© o valor m√≠nimo para heading_accuracy? ").strip()
            jumping = input("Qual √© o valor m√≠nimo para jumping? ").strip()
            strength = input("Qual √© o valor m√≠nimo para strength? ").strip()
            
            
            criterios = {}
            if altura.isdigit():
                criterios['height_cm'] = int(altura)
            if heading_accuracy.isdigit():
                criterios['heading_accuracy'] = int(heading_accuracy)
            if jumping.isdigit():
                criterios['jumping'] = int(jumping)
            if strength.isdigit():
                criterios['strength'] = int(strength)

        elif tipo_defensor in ["RB", "LB"]:
            crossing = input("Qual √© o valor m√≠nimo para crossing? ").strip()
            acceleration = input("Qual √© o valor m√≠nimo para acceleration? ").strip()
            stamina = input("Qual √© o valor m√≠nimo para stamina? ").strip()
            
            
            criterios = {}
            if crossing.isdigit():
                criterios['crossing'] = int(crossing)
            if acceleration.isdigit():
                criterios['acceleration'] = int(acceleration)
            if stamina.isdigit():
                criterios['stamina'] = int(stamina)

        elif tipo_defensor == "GK":
            reflexos = input("Qual √© o valor m√≠nimo para rea√ß√µes (reactions)? ").strip()
            altura = input("Qual √© a altura m√≠nima em cm? ").strip()
            
            
            criterios = {}
            if reflexos.isdigit():
                criterios['reactions'] = int(reflexos)
            if altura.isdigit():
                criterios['height_cm'] = int(altura)

    elif posicao == "Atacante":
        finishing = input("Qual √© o valor m√≠nimo para finaliza√ß√£o (finishing)? ").strip()
        posicao_filtro = "ST|LW|RW"
        if 'ST' in posicao_filtro:
            heading_accuracy = input("Qual √© o valor m√≠nimo para heading_accuracy? ").strip()
            ball_control = input("Qual √© o valor m√≠nimo para ball_control? ").strip()
            
            
            criterios = {}
            if finishing.isdigit():
                criterios['finishing'] = int(finishing)
            if heading_accuracy.isdigit():
                criterios['heading_accuracy'] = int(heading_accuracy)
            if ball_control.isdigit():
                criterios['ball_control'] = int(ball_control)
        else:
            dribbling = input("Qual √© o valor m√≠nimo para dribbling? ").strip()
            acceleration = input("Qual √© o valor m√≠nimo para acceleration? ").strip()
            
            
            criterios = {}
            if finishing.isdigit():
                criterios['finishing'] = int(finishing)
            if dribbling.isdigit():
                criterios['dribbling'] = int(dribbling)
            if acceleration.isdigit():
                criterios['acceleration'] = int(acceleration)

    elif posicao == "Meia":
        posicao_especifica = input("Qual tipo de meia (CM, CAM, CDM, RM, LM)? ").strip().upper()
        posicao_filtro = posicao_especifica
        if posicao_especifica == "CM" or posicao_especifica == "CAM":
            preferred_foot = input("Qual p√© preferido (left, right)? ").strip().capitalize()
            dribbling = input("Qual √© o valor m√≠nimo para dribbling? ").strip()
            long_passing = input("Qual √© o valor m√≠nimo para long_passing? ").strip()
            vision = input("Qual √© o valor m√≠nimo para vision? ").strip()
            
            
            criterios = {}
            if dribbling.isdigit():
                criterios['dribbling'] = int(dribbling)
            if long_passing.isdigit():
                criterios['long_passing'] = int(long_passing)
            if vision.isdigit():
                criterios['vision'] = int(vision)

        elif posicao_especifica == "CDM":
            reactions = input("Qual √© o valor m√≠nimo para reactions? ").strip()
            strength = input("Qual √© o valor m√≠nimo para strength? ").strip()
            aggression = input("Qual √© o valor m√≠nimo para aggression? ").strip()
            interceptions = input("Qual √© o valor m√≠nimo para interceptions? ").strip()
            
            
            criterios = {}
            if reactions.isdigit():
                criterios['reactions'] = int(reactions)
            if strength.isdigit():
                criterios['strength'] = int(strength)
            if aggression.isdigit():
                criterios['aggression'] = int(aggression)
            if interceptions.isdigit():
                criterios['interceptions'] = int(interceptions)

        elif posicao_especifica in ["RM", "LM"]:
            acceleration = input("Qual √© o valor m√≠nimo para acceleration? ").strip()
            dribbling = input("Qual √© o valor m√≠nimo para dribbling? ").strip()
            long_passing = input("Qual √© o valor m√≠nimo para long_passing? ").strip()
            
            
            criterios = {}
            if acceleration.isdigit():
                criterios['acceleration'] = int(acceleration)
            if dribbling.isdigit():
                criterios['dribbling'] = int(dribbling)
            if long_passing.isdigit():
                criterios['long_passing'] = int(long_passing)

    else:
        print("Posi√ß√£o inv√°lida. Por favor, reinicie o programa.")
        return

    age = input("Qual √© a faixa et√°ria do jogador? ").strip()
    overall_rating = input("Qual √© a classifica√ß√£o geral m√≠nima (overall_rating)? ").strip()
    
    
    criterios.update({'age': int(age) if age.isdigit() else None, 'overall_rating': int(overall_rating) if overall_rating.isdigit() else None})

    if age and int(age) < 21:
        potential = input("Qual √© o potencial m√≠nimo (potential)? ").strip()
        criterios.update({'potential': int(potential) if potential.isdigit() else None})

    df_filtrado = df.copy()
    for criterio, valor in criterios.items():
        if valor is not None:
            df_filtrado = aplicar_intervalos(df_filtrado, criterio, valor)

    # Filtros de posi√ß√£o
    jogadores_posicao = jogadores_proximos(df_filtrado, criterios, posicao_filtro)
    
    # Exibindo jogadores encontrados com a fun√ß√£o tabulate
    print(tabulate(jogadores_posicao[['name', 'age', 'positions', 'overall_rating']], headers='keys', tablefmt='pretty'))

arvore_decisao()
```
